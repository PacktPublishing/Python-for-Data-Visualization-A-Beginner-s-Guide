### Getting our first basic line plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [1, 2, 3, 4, 5, 6]

plt.plot(x, y)
plt.show()
```

### Changing Figure Size and Saving the Figure

Using set_figheight() and set_figwidth() 

```python
f = plt.figure()

f.set_figwidth(5)
f.set_figheight(5)

plt.plot(x, y)
plt.show()
```

Method 2: Using figsize

```python

import matplotlib.pyplot as plt
  
plt.figure(figsize=(2, 4)) # width ,hight
display(plt.plot(x, y))
```
